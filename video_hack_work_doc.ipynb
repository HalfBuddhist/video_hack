### 爬取流程一览

提供好用户id, 构造url如下：
http://m.365yg.com/video/app/user/home/?to_user_id=95538078135&iid=29295669458&device_id=14231720144&format=html&app=video_article

获得视频列表，其中元素如下：
http://toutiao.com/group/6540255318931145223/?app=video_article&amp;scheme=snssdk141%3A%2F%2Fdetail%3Fgroupid%3D6540255318931145223

点开后的最终播放地址是：
http://www.365yg.com/a6540255318931145223


页面中的视频地址是通过构造如下的request后得到的json反馈，
http://ib.365yg.com/video/urls/v/1/toutiao/mp4/a519dca5dd6442839c23a73889e56395?r=08633517709597993&s=3290301971&action=GetVideoPlayInfo_YG365A&callback=axiosJsonpCallback1&_=1523763335035
包含在http://s0.pstatp.com/cdn/expire-1-M/byted-player-service/1.0.6/byted-player-service.min.js

其中：
a519dca5dd6442839c23a73889e56395 为视频id, 

r参数为一个随机数，Math.random().toString(10).substring(2)

s参数为这是前面部分地址的CRC32校验值无符号右移0位

其它的参数都可以省略，照样可以得到解析结果，只不过播放地址与带这些参数的有不同。

可以得到response如下：

![Alt text](./img/response.png)

其中main_url就是视频的源地址，base64反编码以后就可以得到视频的播放地址。

地址应该是下面的形式（从you-get --debug中得到）

http://v11-tt.ixigua.com/454323a1197d8fe1d1fe4c84584f2550/5ae4bf3c/video/m/220b5002a2824214c09a0d2562ace2b948b1156ae3a0000af3edb2e6d2e/

Python中下载视频的方法，参见python_learning project，其中还可以得到进度条。

以前的流程中得到的播放地址（非视频地址）是PC端的地址好像(即是列表点开后的地址)，移动端则不相同。

因此，查看是否都可以转化为PC端的地址。

发现有如下规律：

1. 链接地址如：http://toutiao.com/group/6548320003437888007/?app=video_article&scheme=snssdk141%3A%2F%2Fdetail%3Fgroupid%3D6548320003437888007
可以you-get下载，极个别的则不可以，PC与MOBILE地址都是相同的。

1. 移动端转接地址：https://m.365yg.com/group/6547490755353510414/?iid=30578803407&app=video_article&timestamp=1524712859&tt_from=copy_link&utm_source=copy_link&utm_medium=ios&utm_campaign=client_share
PC端会强制跳转（或者说重定向）到
https://www.365yg.com/a6548668527132279303
即上文中说到的地址。
也有极个别的会跳到将m改为www后的地址，如下面的
https://www.365yg.com/group/6547490755353510414/?iid=30578803407&app=video_article&timestamp=1524712859&tt_from=copy_link&utm_source=copy_link&utm_medium=ios&utm_campaign=client_share
其实这个可以通过强制到a开头的地址来下载，只不过得手动，亦有资源，可以发现他们的id是相同的，所以不用模拟301的重定向流程。
后面的两个都可以下，前面的无法下，不如第一种方便。

但是需要cookie.

尝试使用分享后的连接，因为其也可以翻页刷新，分享后的连接是

http://m.365yg.com/video/app/user/home/?to_user_id=4492956276&iid=30578803407&device_id=14231720144&format=html&app=video_article

同样可以使用以下请求来更新列表，用json返回

http://m.365yg.com/video/app/user/home/?to_user_id=52029171784&iid=30578803407&device_id=14231720144&format=json&app=video_article&max_behot_time=1523713438

不同的是max_behot_time字段。

查找该字段的产生规则，简单的上回返回数据的最后一条的"bhot_time"字段前提是json数据的话，如果返回格式是html的话，该字段则会显示在html中，为“data-behot”字段。

上述链接中的format参数，json与html决定了返回内容的格式。

我们肯定用json，经测试其中的article_url与share_url， display_url, url中的地址you-get都可以下载到视频本身, 出于字面意思，我们暂时先用display_url。

至此理论上应该可以得到所有的视频列表，下载的问题也没有问题。


***

简化参数

http://m.365yg.com/video/app/user/home/?to_user_id=4492956276&max_behot_time=1523415689

第一个为user_id, 第二个是视频更新时间，设置为0时为从头开始的约30条数据。

而且这种方法同样可以从video_duration中得到视频的时长，以供过滤。
